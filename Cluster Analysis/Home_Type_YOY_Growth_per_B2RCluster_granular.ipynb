```python
# 1st half of year
first_months = [i for i in range(1, 7)]
compdf1st = compdf[pd.to_datetime(compdf['DateOfObservation']).dt.month.isin(first_months)]

last_months = [i for i in range(7, 13)]
compdf2nd = compdf[pd.to_datetime(compdf['DateOfObservation']).dt.month.isin(last_months)]
compdf2nd
```

```python
yoy1st = pd.DataFrame(compdf1st.groupby(['Characteristics', pd.PeriodIndex(compdf1st['DateOfObservation'], freq="Y")], as_index = True)['Rent'].mean())
yoy1st['Percent Diff %'] = round(yoy1st.pct_change() * 100, 2)
yoy1st.insert(0, 'Half of Year', 1)
yoy1st = yoy1st.reset_index()
#yoy1st.insert(0, 'Characteristics', 'la')
yoy1st
```

```python
char_list = yoy1st['Characteristics'].tolist()

new = yoy1st['Percent Diff %'].tolist()
for i in range(1, len(char_list)):
    if char_list[i] != char_list[i-1]:
        new[i] = None
    else:
        print(char_list[i])
        print(char_list[i+1])
        print('--')
yoy1st['Percent Diff %'] = new
yoy1st
```

```python
yoy2nd = pd.DataFrame(compdf2nd.groupby(['Characteristics',pd.PeriodIndex(compdf2nd['DateOfObservation'], freq="Y")], as_index = True)['Rent'].mean())
yoy2nd['Percent Diff %'] = round(yoy2nd.pct_change() * 100, 2)
yoy2nd.insert(0, 'Half of Year', 2)
yoy2nd = yoy2nd.reset_index()
#yoy2nd.insert(0, 'Characteristics', 'la')
yoy2nd
```
